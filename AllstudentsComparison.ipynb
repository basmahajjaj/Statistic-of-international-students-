```python
#comparison betwwen 3 qara in 2020 ,2021 ,2022
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

!pip install pip install python-bidi
!pip install arabic_reshaper
from bidi.algorithm import get_display

import arabic_reshaper
def reshapingarbic(Text):
    x=arabic_reshaper.reshape(Text)
    x=get_display(x)
    return x

%matplotlib inline
matplotlib.rcParams['interactive'] == True

plt.rcParams['figure.figsize']=(20,20)


labelsstudents = ['2020 year', '2021 year', '2022 year']
labelsstudents = ['Africa', 'Asia', 'europe and america']


x = np.arange(len(labelsstudents))  # the label locations
width = 0.1  # the width of the bars

fig, ax = plt.subplots()
rects2 = ax.bar(x - width/2, qarat2020, width,color='#619ED6', label=reshapingarbic('2020'))
rects1 = ax.bar(x + width/2, qarat2021, width,color='#9CF168', label=reshapingarbic('2021'))
rects3 = ax.bar(x + 1.5*width,qarat2022, width,color='#A31246', label=reshapingarbic('2022'))
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel(reshapingarbic('عدد الطلاب '),fontsize=15)
ax.set_title(reshapingarbic('مقارنه بين عدد الطلاب من حيث القارات في عام 2020,2021,2022'),fontsize=30)
ax.set_xticks(x)
ax.set_xticklabels(labelsstudents,fontsize=30)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
fig.tight_layout()
plt.show()
```
